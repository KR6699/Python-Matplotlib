# Draw a line in a diagram from position (0,0) to position (6,300)

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 300])

plt.plot(xpoints, ypoints)
plt.show()
```

Parameter 1 is an array containing the points on the x-axis.

# Marker Size
You can use the keyword argument markersize or the shorter version, ms to set the size of the markers

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([2, 11, 6, 13, 6])

plt.plot(ypoints, 'o--g', ms = 11)
plt.show()
```

# Subplots

## Display Multiple Plots

With the subplots() function you can draw multiple plots in one figure

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x1 = np.array([2, 11, 6, 13, 6])
y1 = np.array([6, 13, 11, 6, 2])


plt.subplot(1, 2, 1)
plt.title("This is plot 1")
plt.plot(x1,y1, color = 'green', lw = '4')

#plot 2:
x2 = np.array([6, 13, 11, 6, 2])
y2 = np.array([2, 11, 6, 13, 6])

plt.subplot(1, 2, 2)
plt.title("This is plot 2")
plt.plot(x2,y2, color = 'red', lw = '2')

plt.show()
```

The subplots() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x1 = np.array([2, 11, 6, 13, 6])
y1 = np.array([6, 13, 11, 6, 2])


plt.subplot(2, 1, 1)
plt.plot(x1,y1, color = 'green', lw = '4')

#plot 2:
x2 = np.array([6, 13, 11, 6, 2])
y2 = np.array([2, 11, 6, 13, 6])

plt.subplot(2, 1, 2)
plt.plot(x2,y2, color = 'red', lw = '2')

plt.show()
```

# Scatter

## Creating Scatter Plots

With Pyplot, you can use the scatter() function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2, 11, 6, 13, 6])
ypoints = np.array([6, 13, 11, 6, 2])

plt.scatter(xpoints, ypoints)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'green','size':15}

plt.xlabel('x-lebel', fontdict = font2)
plt.ylabel('y-lebel', fontdict = font2)
plt.title("This is Title", fontdict = font1, loc = 'left')
plt.show()
```

## Color Each Dot

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2, 11, 6, 13, 6])
ypoints = np.array([6, 13, 11, 6, 2])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'green','size':15}

colors = np.array(["red","green","blue","yellow","pink"])

plt.xlabel('x-lebel', fontdict = font2)
plt.ylabel('y-lebel', fontdict = font2)
plt.title("This is Title", fontdict = font1, loc = 'left')

plt.scatter(xpoints, ypoints , c=colors)
plt.show()
```

# ColorMap

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

You can specify the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2, 11, 6, 13, 6])
ypoints = np.array([6, 13, 11, 6, 2])


colors = np.array([10, 25, 50, 75, 100])

plt.scatter(xpoints, ypoints , c=colors, cmap='viridis')
plt.show()
```

## You can include the colormap in the drawing by including the plt.colorbar() statement

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2, 11, 6, 13, 6])
ypoints = np.array([6, 13, 11, 6, 2])


colors = np.array([10, 25, 50, 75, 100])

plt.scatter(xpoints, ypoints , c=colors, cmap='viridis')

plt.colorbar()
plt.show()
```

You can adjust the transparency of the dots with the alpha argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis

# Pie Charts

With Pyplot, you can use the pie() function to draw pie charts

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 20, 25, 25])

plt.pie(y)
plt.show() 
```

### Labels

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 20, 25, 25])
l = np.array(['A' , 'B' , 'C' , 'D'])
plt.pie(y , labels = l)
plt.show() 
```

### Start Angle

As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 20, 25, 25])
l = np.array(['A' , 'B' , 'C' , 'D'])
plt.pie(y , labels = l , startangle = 30)
plt.show() 
```

### Explode

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 20, 25, 25])
l = np.array(['A' , 'B' , 'C' , 'D'])
myexplode = [0.1, 0.2, 0.1, 0.3]
plt.pie(y , labels = l , startangle = 30 , explode = myexplode)
plt.show() 
```

### Shadow

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 20, 25, 25])
l = np.array(['A' , 'B' , 'C' , 'D'])
myexplode = [0.1, 0.2, 0.1, 0.3]
plt.pie(y , labels = l , startangle = 30 , explode = myexplode , shadow = True)
plt.show() 
```

### Legend

To add a list of explanation for each wedge, use the legend() function

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 20, 25, 25])
l = np.array(['A' , 'B' , 'C' , 'D'])
plt.pie(y , labels = l)
plt.legend()
plt.show() 
```

### Legend With Header

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 20, 25, 25])
l = np.array(['A' , 'B' , 'C' , 'D'])
plt.pie(y , labels = l)
plt.legend(title = 'Four items:')
plt.show() 
```
